```python
def get_category_type(row):
    try:
        categories_row = row['categories']
    except:
        categories_row = row['venue.categories']
    if len(categories_row) == 0:
        return None
    else:
        return categories_row[0]['name']
```

##### Return significant information for the venues

```python
def getNearbyVenues(name, lat, lng, radius = 500):
    venues = []
    for i, j, k in zip(name, lat, lng):
        print(i)
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            FOURSQ_ID, 
            FOURSQ_SECRET, 
            VERSION, 
            j, 
            k, 
            radius, 
            LIMIT)
        results = requests.get(url).json()["response"]["groups"][0]["items"]
        venues.append([(
            i,
            j,
            k,
            x['venue']['name'],
            x['venue']['location']['lat'],
            x['venue']['location']['lng'],
            x['venue']['id'],
            x['venue']['categories'][0]['name']) for x in results])
    nearby = pd.DataFrame([y for z in venues for y in z])
    nearby.columns = ['County', 
                     'County Latitude', 
                     'County Longitude', 
                     'Venue', 
                     'Venue Latitude', 
                     'Venue Longitude', 
                     'Venue ID', 
                     'Venue Category']
    return nearby
```

```python
num = 5
```

```python
for i in chGrouped['County']:
    print('~~~' + i + '~~~')
    temp = chGrouped[chGrouped['County'] == i].T.reset_index()
    temp.columns = ['venue', 'frequency']
    temp = temp.iloc[1:]
    temp['frequency'] = temp['frequency'].astype(float)
    temp = temp.round({'frequency': 2})
    print(temp.sort_values('frequency', ascending = False).reset_index(drop = True).head(num))
```

```python
def mostCommonVenues(row, num):
    categories = row.iloc[1:]
    categoriesSorted = categories.sort_values(ascending = False)
    return categoriesSorted.index.values[0:num]
```

```python
num = 10
```

```python
indicators = ['st', 'nd', 'rd']
```

##### Columns according to the number of Top Venues

```python
columns = ['County']
for i in np.arange(num):
    try:
        columns.append('{}{} Most Common Venue'.format(i + 1, indicators[i]))
    except:
        columns.append('{}th Most Common Venue'.format(i + 1))
```

##### New DataFrame

```python
countiesVenues = pd.DataFrame(columns = columns)
countiesVenues['County'] = chGrouped['County']
```

```python
for i in np.arange(chGrouped.shape[0]):
    countiesVenues.iloc[i, 1:] = mostCommonVenues(chGrouped.iloc[i, :], num)
```
